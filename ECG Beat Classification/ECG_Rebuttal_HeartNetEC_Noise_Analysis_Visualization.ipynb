# Loading Libraries

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Loading Data

```python
noise_data = pd.ExcelFile("../Noise_Analysis.xlsx")
precision_data = pd.read_excel(noise_data,"Precision")
recall_data = pd.read_excel(noise_data,"Recall")
f1score_data = pd.read_excel(noise_data,"F1-Score")
accuracy_data = pd.read_excel(noise_data,"Accuracy")
```

# Data Visualization

```python
noise_levels = np.arange(0,75,5)
```

```python
for i in ['A','L','N','/','R','V','f','F','!','j','All']:
    plt.figure(figsize=(15,8))
    plt.rc('xtick',labelsize=16)
    plt.rc('ytick',labelsize=16)
    plt.scatter(noise_levels,precision_data[i][:-2],color="red")
    plt.plot(noise_levels,precision_data[i][:-2],color="red",label="Precision")
    
    plt.scatter(noise_levels,recall_data[i][:-2],color="blue")
    plt.plot(noise_levels,recall_data[i][:-2],color="blue",label="Recall")
    
    plt.scatter(noise_levels,f1score_data[i][:-2],color="green")
    plt.plot(noise_levels,f1score_data[i][:-2],color="green",label="F1-Score")
    
    plt.scatter(noise_levels,accuracy_data[i][:-2],color="brown")
    plt.plot(noise_levels,accuracy_data[i][:-2],color="brown",label="Accuracy")
    
    plt.title("Noise Analysis -- Beat Class '"+i+"'", fontsize=24)
    plt.xlabel(r"Noise Level (in dB) $\rightarrow$", fontsize=24)
    plt.ylabel("Metric Value", fontsize=24)
    plt.legend(fontsize=24)
    #plt.savefig("../Class_"+i) if i!='/' else plt.savefig("../Class_slash") 
    plt.show()
```
